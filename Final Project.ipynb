3- Write a function called "to_letter_grade", the function will take the numeric grade from a student and return a letter grade. Test the function by calling it and passing it 79 (you should get back C+). Use the below grading criteria as a reference : 

image.png

```python
def to_letter_grade(grade):
  if grade >= 95:
    return 'A'
  elif grade >= 92 and grade < 95:
    return 'A-'
  elif grade >= 89 and grade < 92:
    return 'B+'
  elif grade >= 84 and grade < 89:
    return 'B'
  elif grade >= 81 and grade < 84:
    return 'B-'
  elif grade >= 78 and grade < 81:
    return 'C+'
  elif grade >= 74 and grade < 78:
    return 'C'
  elif grade >= 70 and grade < 74:
    return 'C-'
  else:
    return 'F'

to_letter_grade(79)
```
